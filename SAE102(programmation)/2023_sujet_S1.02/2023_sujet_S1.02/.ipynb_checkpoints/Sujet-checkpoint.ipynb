## reponse

```python
# chemin du fichier
path = "listing_transactions.csv"

# ouverture fichier
file = open(path, 'r')

# lecture de toute les ligne
rows = file.readlines()

#fermeture fichier
file.close()

# boucle sur chaque transaction et recuperation de la transaction
transactions = []
for row in rows:

    # recuperation de invoiceID, clientID, date et ajout dans le tableau transaction
    transaction = []
    row = row.strip().split(";")
    invoiceID = int(row[0].strip())
    clientID = int(row[1].strip())
    date = row[2].strip()
    transaction.append(invoiceID)
    transaction.append(clientID)
    transaction.append(date)

    # suppresion des info recuper et inutile
    del row[0: 4]

    #parcours du reste de la ligne pour recuperer les items (id, price)
    i = 0
    while i < int(len(row) / 2):
        pos = 2 * i
        item = []
        item.append(int(row[pos].strip()))
        item.append(int(row[pos + 1].strip()))
        transaction.append(item)
        i += 1

    # ajout de la transaction dans la liste des transactions
    transactions.append(transaction)
```

## reponse

```python
# choix d'un client au hasard
transaction = transactions[-1]

#recuperation des information utile a la fracture
date_commande = transaction[2]
numero_fracture = transaction[0]

clientID = transaction[1]
client_name = clients[clientID][0]
client_addresse = clients[clientID][1]

# [nameproduct, quantite, prix_unitaire, prix_total, fournisseur](format des tableau contenu dans product)
# recuperation des info pour chaque produit
products = []
total_ht = 0
i = 3
while i < len(transaction):
    refID = transaction[i][0]
    quantity = transaction[i][1]
    nom_produit = references[refID][2]
    prix_unitaire = references[refID][3]
    prix_total = prix_unitaire * quantity
    total_ht += prix_total
    
    product = [nom_produit, quantity, prix_unitaire, prix_total]
    products.append(product)
    i += 1
    

    # ouverture du fichier en mode 'w' (cree si il n'existe pas)
fracture = open(str(numero_fracture)+ ".txt", "w")

#ecriture et formatage de la facture
fracture.write("Fracture N° " + str(numero_fracture) + "\n")
fracture.write("Date commande : " + date_commande + "\n")
fracture.write("-" * 88 + "\n")
fracture.write("Nom Client : " + client_name + "\n")
fracture.write("Adresse Client : " + client_addresse + "\n")
fracture.write("-" * 40 + " product " + "-" * 40 + "\n")
f = "{0:<30} : {1:<3} * {2:<15} = {3} \n".format("product name", "quantity", "prix unit", "prix total")
fracture.write(f)
for p in products:
    f = "{0:<30} : {1:<8} * {2:<15.2f} = {3:.2f} \n".format(*p)
    fracture.write(f)

fracture.write("-" * 88 + "\n")
fracture.write("Total hors taxe : " + str(total_ht) + "\n")
fracture.write("Total avec taxe : " + str(total_ht*1.2) + "\n")

# close file
fracture.close()

```

## reponse

interface graphique fait avec tkinter qui liste les fichier et permet de voir une facture
       

```python
import tkinter as tk
import os
from typing import Text


win = tk.Tk()
win.geometry("400x400")

var = tk.StringVar(win)

def readfile():
    active = scroll.get(tk.ACTIVE)
    var.set("")
    with open("fractures/"+active, "r") as f:
        lines = f.readlines()

    texte = "".join(lines)
    var.set(texte)


file = sorted(os.listdir("fractures/"))


scroll = tk.Listbox(win, height= 3, width=45)
for i in file:
    scroll.insert(tk.END, i)


button = tk.Button(win, text = " OUVRIR ", command=readfile)

text = tk.Label(win,  textvariable=var)

scroll.grid(row = 0, column= 0, columnspan= 2)
button.grid(row = 0, column=2)
text.grid(row=1, column=0, columnspan=3)

win.mainloop()
```
